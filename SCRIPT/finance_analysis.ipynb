```python
# ============================================================
# PROJET FINAL – MACRO STRATEGY – PRIMARK
# ============================================================
# Contenu :
# 1.1  Seven questions & statements about the company
# 1.2  Porter’s Five Forces – Web Spider (Radar Chart)
# 1.3  Scope – Market situation vs competitors
# 1.4  Relative Position (Cost / Tech)
# 1.5  SWOT + Wordcloud
# 1.6  Opportunities / Risks
# 1.7  Expectations – Income Statement (scénarios)
# ============================================================

# -----------------------------
# Imports
# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud

plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("Set2")

# Pour un style un peu "corporate"
PRIMARY_COLOR = "#003B5C"   # bleu foncé
SECONDARY_COLOR = "#00A1B2" # bleu clair
ACCENT_COLOR = "#F28E2B"    # orange


# ============================================================
# 1.1 – Seven questions and statements about Primark
# ============================================================

questions_statements = pd.DataFrame({
    "Type": [
        "Question", "Question", "Question", "Statement", 
        "Statement", "Question", "Statement"
    ],
    "Content": [
        "How can Primark maintain ultra-low prices while improving sustainability and working conditions in its supply chain?",
        "To what extent can Primark accelerate its digital strategy without cannibalising its store-based business model?",
        "How vulnerable is Primark to economic downturns and changes in consumer sentiment on fast fashion?",
        "Primark’s business model is built on high volume, low margin and strong footfall in physical stores.",
        "Primark has a limited e-commerce presence compared to competitors like H&M and Zara.",
        "How can Primark differentiate beyond price in a highly competitive fast-fashion market?",
        "Primark’s expansion in continental Europe and the US is a key growth driver for the next decade."
    ]
})

print("1.1 – Seven questions and statements about Primark")
display(questions_statements)


# ============================================================
# 1.2 – Porter’s Five Forces – Web Spider (Radar Chart)
# ============================================================
# Scores 1 (faible) à 5 (fort)
# Ces valeurs sont qualitatives, basées sur l’analyse stratégique du secteur fast fashion.

porter_forces = pd.DataFrame({
    "Force": [
        "Threat of New Entrants",
        "Bargaining Power of Suppliers",
        "Bargaining Power of Buyers",
        "Threat of Substitutes",
        "Industry Rivalry"
    ],
    "Score": [2, 3, 4, 3, 5]
})

print("\n1.2 – Porter’s Five Forces – Primark (qualitative scoring)")
display(porter_forces)

# Radar chart (web spider)
labels = porter_forces["Force"].values
scores = porter_forces["Score"].values

# On ferme le cercle
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False)
scores = np.concatenate((scores, [scores[0]]))
angles = np.concatenate((angles, [angles[0]]))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, scores, color=PRIMARY_COLOR, linewidth=2)
ax.fill(angles, scores, color=SECONDARY_COLOR, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10)
ax.set_yticks([1, 2, 3, 4, 5])
ax.set_yticklabels(["1","2","3","4","5"])
ax.set_title("Porter’s Five Forces – Primark (Fast Fashion)", fontsize=14, fontweight="bold", pad=20)
plt.show()


# ============================================================
# 1.3 – Scope – Market situation vs competitors
# ============================================================
# Données simplifiées (ordres de grandeur, en Md€)
# Primark ~ 9–11 Md€, Inditex ~ 35–40 Md€, H&M ~ 20–22 Md€

data_scope = pd.DataFrame({
    "Entreprise": ["Primark", "Inditex (Zara)", "H&M"],
    "Chiffre_affaires_2024_MdEUR": [11.0, 38.5, 21.5],
    "Position_prix": ["Très bas", "Moyen", "Moyen-bas"],
    "Canal_principal": ["Magasins physiques", "Omnicanal", "Omnicanal"],
    "Nombre_magasin_approx": [450, 6000, 4200]
})

print("\n1.3 – Scope – Market situation vs competitors")
display(data_scope)

# Graphique – CA comparé
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

sns.barplot(
    data=data_scope,
    x="Entreprise",
    y="Chiffre_affaires_2024_MdEUR",
    ax=ax[0],
    palette=[PRIMARY_COLOR, SECONDARY_COLOR, ACCENT_COLOR]
)
ax[0].set_title("Chiffre d’affaires 2024 (ordre de grandeur)", fontsize=12, fontweight="bold")
ax[0].set_ylabel("CA 2024 (Md€)")
ax[0].set_xlabel("")

sns.barplot(
    data=data_scope,
    x="Entreprise",
    y="Nombre_magasin_approx",
    ax=ax[1],
    palette=[PRIMARY_COLOR, SECONDARY_COLOR, ACCENT_COLOR]
)
ax[1].set_title("Nombre de magasins (ordre de grandeur)", fontsize=12, fontweight="bold")
ax[1].set_ylabel("Nombre de magasins")
ax[1].set_xlabel("")

plt.suptitle("Primark vs principaux concurrents – Scope de marché", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()


# ============================================================
# 1.4 – Relative Position (Cost / Tech)
# ============================================================
# On place les entreprises sur un plan : 
# X = Niveau de coût (1 = élevé, 5 = très bas coût)
# Y = Maturité technologique / digital (1 = faible, 5 = très avancé)

relative_position = pd.DataFrame({
    "Entreprise": ["Primark", "Inditex (Zara)", "H&M"],
    "Cost_advantage": [5, 3, 4],   # Primark = très bas coût
    "Tech_maturity": [2, 5, 4]     # Primark = peu digital, Zara très avancé
})

print("\n1.4 – Relative Position (Cost / Tech)")
display(relative_position)

fig, ax = plt.subplots(figsize=(7, 6))
for _, row in relative_position.iterrows():
    ax.scatter(row["Cost_advantage"], row["Tech_maturity"], s=200,
               label=row["Entreprise"])

    ax.text(row["Cost_advantage"] + 0.03,
            row["Tech_maturity"] + 0.03,
            row["Entreprise"],
            fontsize=10)

ax.set_xlim(1, 5.2)
ax.set_ylim(1, 5.2)
ax.set_xlabel("Cost advantage (5 = très bas coût)")
ax.set_ylabel("Tech / Digital maturity (5 = très avancé)")
ax.set_title("Relative Position – Cost vs Tech", fontsize=14, fontweight="bold")
plt.grid(True, alpha=0.3)
plt.show()


# ============================================================
# 1.5 – SWOT + Wordcloud
# ============================================================

swot = {
    "Strengths": [
        "Very low prices and strong value perception",
        "High store traffic and impulse buying",
        "Strong brand recognition in Europe",
        "Efficient supply chain and scale purchasing"
    ],
    "Weaknesses": [
        "Limited e-commerce and digital presence",
        "Dependence on physical stores",
        "Exposure to criticism on sustainability and fast fashion",
        "Lower margins compared to premium brands"
    ],
    "Opportunities": [
        "Expansion in US and new European markets",
        "Improving sustainability and circular fashion initiatives",
        "Developing click-and-collect or limited online sales",
        "Partnerships and collaborations with influencers or designers"
    ],
    "Threats": [
        "Intense competition from Zara, H&M, Shein and online players",
        "Regulation on fast fashion and environmental impact",
        "Economic downturn reducing discretionary spending",
        "Rising labour and raw material costs"
    ]
}

# Mettre en DataFrame pour affichage
swot_rows = []
for k, v_list in swot.items():
    for v in v_list:
        swot_rows.append({"Dimension": k, "Item": v})

swot_df = pd.DataFrame(swot_rows)

print("\n1.5 – SWOT – Primark")
display(swot_df)

# Nuage de mots basé sur tous les éléments SWOT
swot_text = " ".join(sum(swot.values(), []))

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    colormap="viridis"
).generate(swot_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("SWOT Wordcloud – Primark", fontsize=14, fontweight="bold")
plt.show()


# ============================================================
# 1.6 – Opportunities / Risks – Tableau + Visualisation
# ============================================================

opportunities_risks = pd.DataFrame({
    "Type": ["Opportunity", "Opportunity", "Opportunity", "Risk", "Risk", "Risk"],
    "Label": [
        "US & new market expansion",
        "Sustainability & circular fashion",
        "Selective digitalisation (click & collect)",
        "Regulation on fast fashion",
        "Competition from ultra-fast fashion (Shein, etc.)",
        "Cost inflation (labour, energy, materials)"
    ],
    "Impact_score_1_5": [5, 4, 3, 4, 5, 4],
    "Probability_1_5": [4, 3, 3, 3, 4, 4]
})

print("\n1.6 – Opportunities / Risks – Primark")
display(opportunities_risks)

# Visualisation – matrice Impact vs Probabilité
fig, ax = plt.subplots(figsize=(8, 6))

for _, row in opportunities_risks.iterrows():
    color = "#2ca02c" if row["Type"] == "Opportunity" else "#d62728"
    ax.scatter(row["Probability_1_5"], row["Impact_score_1_5"],
               s=150, color=color)
    ax.text(row["Probability_1_5"] + 0.03,
            row["Impact_score_1_5"] + 0.03,
            row["Label"],
            fontsize=8)

ax.set_xlim(1, 5.2)
ax.set_ylim(1, 5.2)
ax.set_xlabel("Probability (1–5)")
ax.set_ylabel("Impact (1–5)")
ax.set_title("Opportunities & Risks – Impact vs Probability", fontsize=14, fontweight="bold")
plt.grid(True, alpha=0.3)
plt.show()


# ============================================================
# 1.7 – Expectations – Income Statement (Scénarios)
# ============================================================
# On construit un compte de résultat simplifié pour Primark
# Base : CA 2024 ~ 11 Md€, marge opérationnelle ~ 10% (hypothèse)
# Scénarios sur 3 ans : pessimiste, neutre, optimiste, historique

years = [2024, 2025, 2026, 2027]

# Hypothèses de croissance du CA
growth_rates = {
    "Historical": [0.05, 0.05, 0.05],   # 5% par an
    "Neutral":    [0.04, 0.04, 0.04],   # 4% par an
    "Optimistic": [0.07, 0.07, 0.07],   # 7% par an
    "Pessimistic":[0.01, 0.01, 0.01]    # 1% par an
}

base_revenue_2024 = 11_000  # en millions d'euros (11 Md€)
base_margin = 0.10          # 10% marge opérationnelle

def build_scenario(name, growth_list):
    revenues = [base_revenue_2024]
    for g in growth_list:
        revenues.append(revenues[-1] * (1 + g))
    revenues = np.round(revenues, 1)
    operating_income = np.round(revenues * base_margin, 1)
    return pd.DataFrame({
        "Year": years,
        "Scenario": name,
        "Revenue_MEUR": revenues,
        "Operating_income_MEUR": operating_income
    })

scenario_dfs = []
for scen_name, g_list in growth_rates.items():
    scenario_dfs.append(build_scenario(scen_name, g_list))

income_scenarios = pd.concat(scenario_dfs, ignore_index=True)

print("\n1.7 – Expectations – Income Statement scenarios (Primark)")
display(income_scenarios.head(12))

# Graphique – CA par scénario
plt.figure(figsize=(10, 6))
for scen_name in growth_rates.keys():
    df_s = income_scenarios[income_scenarios["Scenario"] == scen_name]
    plt.plot(df_s["Year"], df_s["Revenue_MEUR"], marker="o", label=scen_name)

plt.title("Primark – Revenue scenarios (2024–2027)", fontsize=14, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Revenue (M€)")
plt.legend(title="Scenario")
plt.grid(True, alpha=0.3)
plt.show()

# Graphique – Résultat opérationnel par scénario
plt.figure(figsize=(10, 6))
for scen_name in growth_rates.keys():
    df_s = income_scenarios[income_scenarios["Scenario"] == scen_name]
    plt.plot(df_s["Year"], df_s["Operating_income_MEUR"], marker="o", label=scen_name)

plt.title("Primark – Operating income scenarios (2024–2027)", fontsize=14, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Operating income (M€)")
plt.legend(title="Scenario")
plt.grid(True, alpha=0.3)
plt.show()


# ============================================================
# BONUS – ROI & Barrières à l’entrée (texte structuré)
# ============================================================
# Tu peux garder cette partie en texte dans ton rapport, ou la transformer
# en DataFrame + affichage si tu veux tout faire en “mode data”.

roi_barriers = pd.DataFrame({
    "Dimension": [
        "ROI drivers",
        "ROI drivers",
        "ROI drivers",
        "Barriers to entry",
        "Barriers to entry",
        "Barriers to entry"
    ],
    "Item": [
        "High store productivity and footfall in prime locations",
        "Low operating costs per unit thanks to scale and simplified operations",
        "Fast inventory rotation and limited markdowns",
        "Need for strong purchasing power and scale in sourcing",
        "Access to prime retail locations in city centres and malls",
        "Brand recognition and trust on price/quality ratio"
    ]
})

print("\nBONUS – ROI & Barrières à l’entrée – Primark")
display(roi_barriers)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, Circle

# ============================
# SWOT CONTENT (ENGLISH)
# ============================

strengths = [
    "Ultra-low cost structure",
    "High store productivity",
    "Strong value-for-money brand",
    "Efficient high-volume supply chain"
]

weaknesses = [
    "Very limited digital presence",
    "No e-commerce capabilities",
    "Dependence on physical stores",
    "Sustainability reputation concerns"
]

opportunities = [
    "Digital expansion (click & collect)",
    "Geographic expansion",
    "Sustainability initiatives",
    "Growing demand for affordable fashion"
]

threats = [
    "Intense competition (Zara, H&M, Shein)",
    "Rise of second-hand platforms",
    "Regulatory pressure on sustainability",
    "Rising production costs in Asia"
]

# ============================
# DESIGN SETTINGS
# ============================

BG = "#F4F6F8"
NAVY = "#0A1A2F"

COLOR_STRENGTHS = "#27AE60"      # Green
COLOR_WEAKNESSES = "#C0392B"     # Red
COLOR_OPPORTUNITIES = "#2980B9"  # Blue
COLOR_THREATS = "#D35400"        # Orange

SHADOW = "#00000030"

plt.rcParams["font.family"] = "DejaVu Sans"
plt.rcParams["font.size"] = 11

# ============================
# FIGURE
# ============================

fig, ax = plt.subplots(figsize=(14, 9))
fig.patch.set_facecolor(BG)
ax.set_facecolor(BG)
ax.axis("off")

# ============================
# FUNCTION TO DRAW A ROUNDED SWOT BOX
# ============================

def swot_box(x, y, w, h, color, title, items):
    # Shadow
    shadow = FancyBboxPatch(
        (x+0.015, y-0.015), w, h,
        boxstyle="round,pad=0.03,rounding_size=30",
        linewidth=0, facecolor=SHADOW, zorder=1
    )
    ax.add_patch(shadow)

    # Main box
    box = FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.03,rounding_size=30",
        linewidth=0, facecolor=color, zorder=2
    )
    ax.add_patch(box)

    # Decorative circle
    circ = Circle((x + w - 0.08, y + h - 0.08), 0.06,
                  color="white", alpha=0.25, zorder=3)
    ax.add_patch(circ)

    # Title
    ax.text(x + 0.04, y + h - 0.13, title,
            fontsize=17, fontweight="bold", color="white")

    # Bullet list
    y_text = y + h - 0.25
    for item in items:
        ax.text(x + 0.06, y_text, "• " + item,
                fontsize=11.5, color="white", wrap=True)
        y_text -= 0.11

# ============================
# DRAW THE 4 SWOT BOXES (4 DISTINCT COLORS)
# ============================

swot_box(0.05, 0.55, 0.40, 0.35, COLOR_STRENGTHS, "STRENGTHS", strengths)
swot_box(0.55, 0.55, 0.40, 0.35, COLOR_WEAKNESSES, "WEAKNESSES", weaknesses)
swot_box(0.05, 0.10, 0.40, 0.35, COLOR_OPPORTUNITIES, "OPPORTUNITIES", opportunities)
swot_box(0.55, 0.10, 0.40, 0.35, COLOR_THREATS, "THREATS", threats)

# ============================
# TITLE
# ============================

plt.text(0.5, 0.94, "SWOT Analysis – Primark",
         fontsize=26, fontweight="bold",
         color=NAVY, ha="center")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np   # <-- IMPORTANT : nécessaire pour cos/sin

# ============================
# DATA
# ============================

labels = ["H1 (0–6 months)", "H2 (6–18 months)", "H3 (18–36 months)"]
durations = [6, 12, 18]

# ============================
# COLORS (3 DISTINCT, PREMIUM)
# ============================

colors = [
    "#247BA0",  # Deep Blue
    "#70C1B3",  # Aqua Green
    "#F25F5C"   # Coral Red
]

# ============================
# FIGURE
# ============================

plt.figure(figsize=(11, 9))
plt.rcParams["font.family"] = "DejaVu Sans"

# ============================
# PIE CHART (DONUT + EXTERNAL LABELS)
# ============================

wedges, _ = plt.pie(
    durations,
    colors=colors,
    startangle=90,
    wedgeprops={"linewidth": 2, "edgecolor": "white"}
)

# ============================
# LABELS EXTERNES + LIGNES
# ============================

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1) / 2 + p.theta1
    x = 1.25 * np.cos(np.deg2rad(ang))
    y = 1.25 * np.sin(np.deg2rad(ang))

    plt.annotate(
        labels[i],
        xy=(np.cos(np.deg2rad(ang)), np.sin(np.deg2rad(ang))),
        xytext=(x, y),
        fontsize=15,
        fontweight="bold",
        color="#0A1A2F",
        ha="center",
        va="center",
        arrowprops=dict(arrowstyle="-", lw=2, color="#555")
    )

# ============================
# DONUT EFFECT
# ============================

centre_circle = plt.Circle((0, 0), 0.55, fc="#F4F6F8")
plt.gca().add_artist(centre_circle)

# ============================
# TITLE
# ============================

plt.title(
    "Strategic Roadmap Distribution (H1 – H2 – H3)",
    fontsize=26,
    fontweight="bold",
    color="#0A1A2F",
    pad=30
)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ============================
# DONNÉES DES RISQUES
# ============================

risks = [
    "Digital service inconsistency",
    "Supply chain complexity",
    "Capex overruns",
    "Cybersecurity vulnerabilities",
    "Regulatory non-compliance",
    "Brand reputation damage"
]

prob = [4, 3, 3, 5, 4, 4]   # Probabilité (1–5)
impact = [4, 5, 3, 5, 5, 4] # Impact (1–5)

data = np.array([prob, impact]).T  # shape (6, 2)

# ============================
# FIGURE
# ============================

plt.figure(figsize=(10, 6))
plt.rcParams["font.family"] = "DejaVu Sans"

# Heatmap
im = plt.imshow(data, cmap="OrRd", aspect="auto")

# Axes
plt.xticks([0, 1], ["Probability", "Impact"], fontsize=12, fontweight="bold")
plt.yticks(range(len(risks)), risks, fontsize=11)

# Valeurs dans les cases
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        plt.text(j, i, data[i, j],
                 ha="center", va="center",
                 color="black", fontsize=12, fontweight="bold")

# Barre de couleur
cbar = plt.colorbar(im)
cbar.set_label("Risk level (1–5)", fontsize=12, fontweight="bold")

plt.title("Risk Matrix – Probability & Impact (Heatmap)",
          fontsize=16, fontweight="bold", pad=15)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
import random

# 1. DONNÉES ENRICHIES (Catégorisées pour les couleurs)
# Format : { "Mot": (Poids, "Catégorie") }
data_strat = {
    # STRENGTHS (Blue)
    'Cost Leadership': (100, 'S'), 'No Frills': (85, 'S'), 'Volume Sales': (90, 'S'),
    'Prime Locations': (85, 'S'), 'ABF Group': (80, 'S'), 'Trend Agility': (75, 'S'),
    'Scale Economies': (80, 'S'), 'Logistics': (65, 'S'), 'Brand Equity': (70, 'S'),
    
    # WEAKNESSES (Red/Grey)
    'No Home Delivery': (95, 'W'), 'Store Reliance': (80, 'W'), 'Thin Margins': (85, 'W'),
    'Disposable Image': (70, 'W'), 'LFL Pressure': (65, 'W'), 'Digital Lag': (75, 'W'),
    
    # OPPORTUNITIES (Green)
    'US Expansion': (95, 'O'), 'Click & Collect': (90, 'O'), 'Primark Cares': (85, 'O'),
    'Circular Design': (75, 'O'), 'India Partnership': (70, 'O'), 'Athleisure': (65, 'O'),
    'Home Decor': (60, 'O'), 'AI Forecasting': (75, 'O'), 'Recycled Fibres': (70, 'O'),
    
    # THREATS (Orange/Yellow)
    'Shein Rivalry': (95, 'T'), 'Temu Attack': (90, 'T'), 'Wage Inflation': (85, 'T'),
    'Greenwashing Law': (80, 'T'), 'Red Sea Crisis': (75, 'T'), 'Cotton Prices': (70, 'T'),
    'Labour Costs': (80, 'T'), 'Fast Fashion Ban': (75, 'T')
}

# Préparation des fréquences pour WordCloud
frequencies = {k: v[0] for k, v in data_strat.items()}

# 2. LOGIQUE DE COULEUR PAR SEGMENT
def strategic_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    category = data_strat.get(word, (0, 'S'))[1]
    colors = {
        'S': ["#004c76", "#0078b7", "#00a8e0"], # Blues
        'W': ["#d55c4b", "#9e3d31", "#6d2a22"], # Reds
        'O': ["#2e8b57", "#3cb371", "#20603d"], # Greens
        'T': ["#f08a24", "#d47216", "#a35811"]  # Oranges
    }
    return random.choice(colors.get(category, ["#000000"]))

# 3. CRÉATION DU NUAGE "HIGH DENSITY"
# On augmente le scale et on ajuste le ratio horizontal pour le look "pro"
wc = WordCloud(
    background_color='white',
    width=1920,
    height=1080,
    max_words=300,
    prefer_horizontal=0.9,
    relative_scaling=0.4,
    color_func=strategic_color_func,
    repeat=False
).generate_from_frequencies(frequencies)

# 4. AFFICHAGE ET SAUVEGARDE
plt.figure(figsize=(20, 10), dpi=100)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)

# Exportation pour votre PowerPoint
wc.to_file("primark_strategic_overview.png")
print("✅ Nuage de mots stratégique généré avec succès !")
plt.show()
```

```python
import matplotlib.pyplot as plt

brands = {
    "Primark": (2, 2),
    "Zara": (8, 8),
    "Uniqlo": (7, 7),
    "H&M": (5, 6),
    "Action": (1.5, 1.5),
    "Lidl textile": (1.8, 1.8),
}

plt.figure(figsize=(10, 8))

for b, (wtp, value) in brands.items():
    plt.scatter(wtp, value, s=120, color="black" if b != "Primark" else "red")
    plt.text(wtp + 0.1, value + 0.1, b, fontsize=11, color="red" if b == "Primark" else "black")

plt.xlabel("Willingness To Pay (relative)", fontsize=13)
plt.ylabel("Perceived Value / Price Level", fontsize=13)
plt.title("WTP Matrix – Primark vs Competitors", fontsize=16, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(10, 6))

# Création des blocs de la matrice
rects = [
    (0, 1, 1, 1, '#0078b7', 'COST\nLEADERSHIP'),
    (1, 1, 1, 1, '#e0301e', 'DIFFERENTIATION'),
    (0.5, 0.5, 1, 0.5, '#f08a24', 'STUCK IN THE MIDDLE'),
    (0, 0, 1, 0.5, '#f5a623', 'COST FOCUS'),
    (1, 0, 1, 0.5, '#2e8b57', 'DIFFERENTIATION\nFOCUS')
]

for x, y, w, h, color, text in rects:
    ax.add_patch(patches.Rectangle((x, y), w, h, facecolor=color, edgecolor='white', linewidth=2))
    ax.text(x + w/2, y + h/2, text, color='white', ha='center', va='center', fontweight='bold')

# Positionnement de Primark
ax.text(0.3, 1.5, 'PRIMARK', bbox=dict(facecolor='white', edgecolor='#0078b7', boxstyle='round,pad=0.5'),
        fontsize=12, fontweight='bold', color='#0078b7')

ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
ax.set_xticks([0.5, 1.5])
ax.set_xticklabels(['LOW COST', 'DIFFERENTIATION'])
ax.set_yticks([0.25, 1.5])
ax.set_yticklabels(['NICHE MARKET', 'TOTAL MARKET'])
ax.set_ylabel('Scope', fontweight='bold')
ax.set_title('4. Position of Primark: Competitive Advantage', fontsize=14, pad=20)
plt.show()
```
